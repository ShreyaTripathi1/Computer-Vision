# Morphological operation (opening, closing, erosion, dilation)

Morphological operations are techniques in image processing that process images based on their shapes. These operations are typically used to remove noise, separate objects, or emphasize structures in an image.

The four common morphological operations are:

- **Erosion:** Removes pixels at the boundaries of the foreground object.
- **Dilation:** Adds pixels to the boundaries of the foreground object.
- **Opening:** Erosion followed by dilation; removes small objects from the background.
- **Closing:** Dilation followed by erosion; removes small holes from the foreground.

```python
# 22BAI10033
# Shreya Tripathi
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the original image (color)
img = cv2.imread('lotus.jpg')

# Convert the original image to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Define a kernel (structuring element) for morphological operations
kernel = np.ones((5, 5), np.uint8)

# Apply Erosion
erosion = cv2.erode(img_gray, kernel, iterations=1)

# Apply Dilation
dilation = cv2.dilate(img_gray, kernel, iterations=1)

# Apply Opening (erosion followed by dilation)
opening = cv2.morphologyEx(img_gray, cv2.MORPH_OPEN, kernel)

# Apply Closing (dilation followed by erosion)
closing = cv2.morphologyEx(img_gray, cv2.MORPH_CLOSE, kernel)

# Display the results using matplotlib
plt.figure(figsize=(12, 6))

# Original Image (colored)
plt.subplot(2, 3, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for proper display
plt.axis('off')

# Erosion
plt.subplot(2, 3, 2)
plt.title('Erosion')
plt.imshow(erosion, cmap='gray')
plt.axis('off')

# Dilation
plt.subplot(2, 3, 3)
plt.title('Dilation')
plt.imshow(dilation, cmap='gray')
plt.axis('off')

# Opening
plt.subplot(2, 3, 4)
plt.title('Opening')
plt.imshow(opening, cmap='gray')
plt.axis('off')

# Closing
plt.subplot(2, 3, 5)
plt.title('Closing')
plt.imshow(closing, cmap='gray')
plt.axis('off')

# Show the images
plt.show()
```
