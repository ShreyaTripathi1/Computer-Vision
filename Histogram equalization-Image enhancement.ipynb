# Histogram equalization-Image enhancement

Histogram equalization is a technique used in image processing to improve the contrast of an image by adjusting its intensity distribution. The goal is to enhance the image such that the histogram of the pixel intensities is more evenly spread across the available intensity range, resulting in a better contrast.

```python
# SHREYA TRIPATHI
# 22BAI10033

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image in grayscale
img = cv2.imread('lotus.jpg', 0)

# Apply CLAHE (Contrast Limited Adaptive Histogram Equalization)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
img_clahe = clahe.apply(img)

# Apply Histogram Equalization
equ = cv2.equalizeHist(img)

# Stack original and equalized images side by side
res = np.hstack((img, equ))

# Display the images using Matplotlib
plt.figure(figsize=(12, 6))

# Original and Histogram Equalized Images
plt.subplot(1, 2, 1)
plt.title("Original and Equalized")
plt.axis('off')
plt.imshow(res, cmap='gray')

# CLAHE Image
plt.subplot(1, 2, 2)
plt.title("CLAHE Image")
plt.axis('off')
plt.imshow(img_clahe, cmap='gray')

# Show the plot
plt.show()

# Save the CLAHE image
cv2.imwrite('clahe.jpg', img_clahe)
```
